```python
# Where it will ask for user input
experience = int(input("Enter number of years of experience: "))
age = int(input("Enter age: "))

# Determine ATR
if experience > 25 and age >= 55:
    atr = 5600000
elif experience > 20 and age >= 45:
    atr = 4480000
elif experience > 10 and age >= 35:
    atr = 1500000
elif experience < 10 and age < 35:
    atr = 550000
else:
    atr = 0

# print result
print("Experience: ", experience)
print("Age: ", age)
print("Annual Tax Revenue (ATR): N", atr)
```
