```python
import tkinter as tk
import csv

# Function for submit button
def submit():
    name = name_entry.get().lower()
    dept = dept_entry.get().lower()

    found = False

    with open("GIG-logistics.csv", "r") as file:
        reader = csv.reader(file)

    for row in reader:
        file_name = row[1].lower()      # Column B
        file_dept = row[2].lower()      # Column C

        if name == file_name and dept == file_dept:
          found = True

    if found:
        result_label.config(text="Welcome " + name.title())
    else:
        result_label.config(text="Employee does not exist or wrong department")


# Create window
root = tk.Tk()
root.title("Employee Verification Form")
root.geometry("400x250")

# Title
title = tk.Label(root, text="Employee Verification Form", font=("Arial",16))
title.pack(pady=15)

# Name
name_label = tk.Label(root, text="Employee Name")
name_label.pack()

name_entry = tk.Entry(root, width=35)
name_entry.pack(pady=5)

# Department
dept_label = tk.Label(root, text="Department")
dept_label.pack()

dept_entry = tk.Entry(root, width=35)
dept_entry.pack(pady=5)

# Submit button
submit_button = tk.Button(root, text="SUBMIT", command=submit)
submit_button.pack(pady=15)

# Result text
result_label = tk.Label(root, text="")
result_label.pack()

# Run program
root.mainloop()
```
